Решение а)

Для системы с $n$ уравнениями и $n$ неизвестными, если система уравнений невырождена (то есть $detA\neq 0$), то система определена, то есть имеет единственное решение, и это решение может быть найдено по формулам Крамера:

$$x_{i}=\frac{detA_{i}}{detA},$$

где $A_{i}$ — это матрица, получаемая заменой $i$-го столбца на вектор-столбец свободных членов $b$.

$detA=\begin{vmatrix}
1 & -2\\ 
3 & -4
\end{vmatrix}=1 \cdot (-4)-3\cdot (-2) =-4+6 = 2\neq 0,$ система совместна

$detA_{1}=\begin{vmatrix}
1 & -2\\ 
7 & -4
\end{vmatrix}=1 \cdot (-4)-7\cdot (-2) =-4+14 = 10$

$detA_{2}=\begin{vmatrix}
1 & 1\\ 
3 & 7
\end{vmatrix}=1 \cdot 7-3\cdot 1 =7-3 = 4$

$x_{1} = \frac{detA_{1}}{detA} = \frac{10}{2}=5$

$x_{2} = \frac{detA_{2}}{detA} = \frac{4}{2}=2$

б) $\begin{cases}
2x_{1}-x_{2}+5x_{3}=10 \\
x_{1}+x_{2}-3x_{3}=-2 \\
2x_{1}+4x_{2}+x_{3}=1

\end{vmatrix}= 
2\begin{vmatrix}

4 & 1
\end{vmatrix}+
\begin{vmatrix}
1 & -3\\ 
2 & 1
\end{vmatrix}+
5\begin{vmatrix}
1 & 1\\ 
2 & 4
\end{vmatrix}=2(1+12)+(1+6)+5(4-2)=26+7+10=43\neq 0
$


$detA_{1}=\begin{vmatrix}
10 & -1 & 5\\ 
-2 & 1 & -3 \\
1 & 4 & 1
\end{vmatrix}= 
10\begin{vmatrix}
1 & -3\\ 
4 & 1
\end{vmatrix}+
\begin{vmatrix}
-2 & -3\\ 
1 & 1
\end{vmatrix}+
5\begin{vmatrix}
-2 & 1\\ 
1 & 4
\end{vmatrix}=10(1+12)+(-2+3)+5(-8-1)=130+1-45=86
$

$x_{1} = \frac{detA_{1}}{detA} = \frac{86}{43}=2$

$detA_{2}=\begin{vmatrix}
2 & 10 & 5\\ 
1 & -2 & -3 \\
2 & 1 & 1
\end{vmatrix}=

\end{vmatrix}=2(-2+3)-10(1+6)+5(1+4)=2-70+24=-43
$

$x_{2} = \frac{detA_{2}}{detA} = \frac{-43}{43}=-1$

$detA_{3}=\begin{vmatrix}

\end{vmatrix}=

$

$x_{3} = \frac{detA_{3}}{detA} = \frac{43}{43}=1$

```python
def kramer(A, b):
    det_A = np.linalg.det(A)
    print(f'detA равен = {det_A}')
    result = []
    if np.abs(det_A) > 0:
        for i in range(A.shape[0]):
            A_i = np.hstack((A[:, :i], b, A[:, i+1:]))
            result.append(np.linalg.det(A_i) / det_A)
        return np.round(result, 2)
    else:
        return 'det A равен нулю'
```

```python
A = np.array([[1, -2], 
              [3, -4]])
B = np.array([[1, 7]]).T
print(f'Решение а): {kramer(A, B)}')
```

```python
A = np.array([[2, -1, 5], 
              [1, 1, -3], 
              [2, 4, 1]])
b = np.array([[10, -2, 1]]).T
print(f'Решение б): {kramer(A, b)}')
```

```python
def LU(A):
    assert A.shape[0] == A.shape[1] and A[0][0] !=0 and np.abs(np.linalg.det(A)) > 0.000001
    L = np.zeros_like(A)
    U = np.zeros_like(A)    
    U[0] = A[0]
    L[:, 0] = A[:, 0] / U[0, 0]
    for i in range(A.shape[1]):
        for j in range(i, A.shape[1]):
            U[0, i] = A[0, i]
            L[i, 0] = A[i, 0] / U[0, 0]
            u_sum, l_sum = 0, 0
            for k in range(i):
                u_sum += L[i, k] * U[k, j]
                l_sum += L[j, k] * U[k, i]
            U[i, j] = A[i, j] - u_sum
            L[j, i] = (A[j, i] - l_sum) / U[i, i]
    return U, L
```

```python
A = np.array([[1, 2, 4], 
              [2, 9, 12], 
              [3, 26, 30]])
U, L = LU(A)
print(f'A: \n{A}')
print(f'U: \n{U}')
print(f'L: \n{L}')
print(f'Проверка LU = \n{L.dot(U)}')
if np.array_equal (A,L.dot(U)):
    print('А = LU, значит решение верно')
else:
    print('Error')
```
